```python
import os
import pandas as pd

folder_path = './Dataset'
data = []

for filename in os.listdir(folder_path):
    if filename.endswith('.txt') :  
        file_path = os.path.join(folder_path, filename)
        
        region_name = filename.split('_')[0] if '_' in filename else filename.split('.')[0]
        
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            
            for line in lines:
                name = line.strip()  
                if name:  
                    surname = name.split()[-1]  
                    data.append({"Surname": surname, "Region": region_name})

df = pd.DataFrame(data)
df['Region'] = df['Region'].str.replace(r'\d+', '', regex=True).str.strip()

output_path = 'surname.csv'
df.to_csv(output_path, index=False)
print(f"Surnames and country names saved to {output_path}")
```
