```python
#Featues
Xfeatures = df['Surname'].fillna(" ").astype(str).tolist()
ylabels = df['Region']
```

```python
cv = CountVectorizer()
X = cv.fit_transform(Xfeatures)
```

```python
print(X.shape[0])       # Check the number of samples in X
print(len(ylabels))     # Check the length of ylabels
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, ylabels,test_size = 0.2, random_state=50)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
#Support Vector Machine (SVM) 
```

```python
svm_model = SVC(kernel='linear', random_state=50)  

svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix using seaborn
plt.figure(figsize=(10, 5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", cbar=False)
plt.title("Confusion Matrix (SVM)")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
nb_model = MultinomialNB()

nb_model.fit(X_train, y_train)

y_pred = nb_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)

fig = plt.figure(figsize=(10, 5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.title("Confusion Matrix (Multinomial Naive Bayes)")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#decision Tree Implementation:
```

```python
dt_model = DecisionTreeClassifier(random_state=50)

dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)

fig = plt.figure(figsize=(10, 5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.title("Confusion Matrix (Decision Tree)")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
dt_model = RandomForestClassifier(n_estimators=100,random_state=50)

dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)

fig = plt.figure(figsize=(10, 5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.title("Confusion Matrix (Random Forest Tree)")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
from sklearn.ensemble import VotingClassifier

# Combine models into a voting classifier
voting_clf = VotingClassifier(estimators=[
    ('nb', nb_model),
    ('rf', dt_model),
    ('svm',svm_model),
], voting='hard')  # Use 'hard' for majority voting, 'soft' for averaging probabilities

# Train the voting classifier
voting_clf.fit(X_train, y_train)

# Make predictions
y_pred = voting_clf.predict(X_test)

# Evaluate the performance
print("Voting Classifier Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from joblib import dump, load

# Save the trained VotingClassifier model to a file
dump(voting_clf, 'voting_classifier.pkl')
```
